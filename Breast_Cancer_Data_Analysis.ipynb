## Splitting the Data into equal halves

```python
X_train, X_test, y_train, y_test = train_test_split(df, df_y, test_size=0.5, random_state=0)
```

```python
print(len(X_train), len(X_test))
```

## Logistic Regression using default parameters

```python
#importing the required libraries
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#Training the model
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
#Calculating the confusion matrix, accuracy, precision, recall, f1 score (using micro, macro and weighted as shown below) 
pred = lr.predict(X_test)
print("F1Score: ", metrics.f1_score(y_test, pred, average='weighted'))
print("Accuracy Score", metrics.accuracy_score(y_test, pred))
print(metrics.confusion_matrix(y_test, pred))
print(metrics.classification_report(y_test, pred))
```

### So we get an accuracy of 0.9543 using above model

## SVM using Linear Kernel with Default Value

```python
#Training the model using Support Vector Machine with Linear Kernel
#svc_linear_kernel = svm.SVC(kernel="linear",C=10,gamma='auto')
svc_linear_kernel = svm.SVC(kernel="linear")
svc_linear_kernel.fit(X_train, y_train)
svc_linear_pred = svc_linear_kernel.predict(X_test)

print("Accuracy Score", metrics.accuracy_score(y_test, svc_linear_pred))
print(metrics.confusion_matrix(y_test, svc_linear_pred))
print(metrics.classification_report(y_test, svc_linear_pred))
```

### So we get an accuracy of 0.9578 using SVM using Linear Kernel 

## SVM using RBF Kernel with Default Values

```python
#svc_rbf_kernel = svm.SVC(kernel="rbf",C=1,gamma='auto')
svc_rbf_kernel = svm.SVC(kernel="rbf")
svc_rbf_kernel.fit(X_train, y_train)
svc_rbf_pred = svc_linear_kernel.predict(X_test)

print("Accuracy Score", metrics.accuracy_score(y_test, svc_rbf_pred))
print(metrics.confusion_matrix(y_test, svc_rbf_pred))
print(metrics.classification_report(y_test, svc_rbf_pred))
```

### So we get an accuracy of 0.9578 using SVM using Linear Kernel,
### It is also find that accuracy (for this dataset and the given split ) remains constact while using default values for both SVM with Linear and RBF Kernel. But as compared to logistic regression there is improvement in the accuracy score for SVM models.
### Also, I have plotted precison, recall and f1 score for each model and it is found that these score improves(comes to 96%) for SVM model as whole compared to Logistic Regression model where these scores were mostly around 95%.

## StandardScalar

```python
# The idea behind StandardScaler is that it will transform  data such that its distribution will
#have a mean value 0 and standard deviation of 1. Given the distribution of the data, each value 
#in the dataset will have the sample mean value subtracted, and 
#then divided by the standard deviation of the whole dataset.
# Importing required libraries from sklearn and scaling the data using StandardScalar¶
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data_train = scaler.fit_transform(X_train)
scaled_data_test = scaler.fit_transform(X_test)
```

### Once data is scaled using standardscalar, using these data valus with the models trained and tested above

### Logisitic Regression with default values and scaled data

```python
lr.fit(scaled_data_train, y_train)
pred = lr.predict(scaled_data_test)

print("Accuracy Score", metrics.accuracy_score(y_test, pred))
print(metrics.confusion_matrix(y_test, pred))
print(metrics.classification_report(y_test, pred))
```

### It can be seen here that on using scaled data with logistic regression, there is improvement in the accuracy score as well as other scores displayed above, when compared with the logistic model used with default values and unscaled data(0.9543). Here the accuracy score comes to 0.9614.

### SVM using Linear Kernel with default values and scaled data

### SVM using RBF Kernel  with default values and scaled data

## GridSearchCV
##### Grid search is the process of performing hyper parameter tuning in order to determine the optimal values for a given model. This is significant as the performance of the entire model is based on the hyper parameter values specified.

### Logisitic Regression
#### Tuning the C parameter values

```python
#importing libraries
from sklearn.model_selection import GridSearchCV
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#creating the model, traing and evaluting it
gsc = GridSearchCV(
        estimator=lr,
        param_grid={'C': [0.1, 1, 100, 1000]}, cv=5)
gs = gsc.fit(X_train, y_train)
gs.predict(X_test)
```

```python
#Showing the results for various C parameter values taken
gs.cv_results_
```

#### Visualizing the accuracy as function of the parameters

```python
#ploting the above obtained values using pandas dataframe for better visualization
results = pd.DataFrame(gs.cv_results_)
results
```

###  Plotting the mean_test_score obtained for different 'C' using graph plot and looking for best performing model

```python
label = list(range(0, len(results['params'])))
plt.title('Models Comparison for various Configuration')
plt.xlabel('Different Models Created as in Table')
plt.ylabel('Accuracy')
plt.plot(label, results['mean_test_score'])
plt.xticks(label)
plt.show()
```

### So out of 4 'C' Values taken for hyper parameter tuning in order to determine the optimal values for a given model, we see that mean_test_score is highest for C=2 and is same for C=2 and C=3. The score is 0.950704 which is somewhat lower than the Logistic Regression created and evaluated above using scaled and unscaled data, also this value is lowest of all the model created above using scaled and uncsaled data.

## Coefficient of LogisticRegression

```python
import numpy as np
```

```python
#Of all the Coefficient for various LogisticRegression model created, I have shown the coeffificet for best model created
gs.best_estimator_.coef_
```

```python
#Finding the feature importance score
feature_importance = abs(gs.best_estimator_.coef_[0])
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

featfig = plt.figure(figsize=(10,6))
featax = featfig.add_subplot(1, 1, 1)
featax.barh(pos, feature_importance[sorted_idx], align='center')
featax.set_yticks(pos)
featax.set_yticklabels(np.array(X_train.columns)[sorted_idx], fontsize=10)
featax.set_xlabel('Feature Importance Score')

# plt.tight_layout()   
plt.show()
```

###  What does coefficient say about faetures and Role they play: A regression coefficient describes the size and direction of the relationship between a predictor and the response variable. Variables of interest in an experiment (those that are measured or observed) are called response or dependent variables. Other variables in the experiment that affect the response and can be set or measured by the experimenter are called predictor.
### Coefficients are the numbers by which the values of the term are multiplied in a regression equation. We use the coefficient to determine whether a change in a predictor variable makes the event more likely or less likely. The estimated coefficient for a predictor represents the change in the link function for each unit change in the predictor, while the other predictors in the model are held constant. The relationship between the coefficient and the probability depends on several aspects of the analysis, including the reference event for the response and the reference levels for categorical predictors. 

### Which one are important: Generally, positive coefficients make the event more likely and negative coefficients make the event less likely. An estimated coefficient near 0 implies that the effect of the predictor is small. Here we see above, that we have a mix of positive and negative coefiicients.

```python
feature_importance = abs(gs1.best_estimator_.coef_[0])
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

featfig = plt.figure(figsize=(10,6))
featax = featfig.add_subplot(1, 1, 1)
featax.barh(pos, feature_importance[sorted_idx], align='center')
featax.set_yticks(pos)
featax.set_yticklabels(np.array(X_train.columns)[sorted_idx], fontsize=10)
featax.set_xlabel('Feature Importance Score')

# plt.tight_layout()   
plt.show()
```

### What does coefficient say about faetures: In Linear SVM, the result is a hyperplane that separates the classes as best as possible. The weights represent this hyperplane, by giving the coordinates of a vector which is orthogonal to the hyperplane - these weights are the coefficients given by svm.coef_. 
### Role they play: The direction of the vector thus created gives us the predicted class, so if we take the dot product of any point i.e coefficient with the vector, we can tell on which side it is: if the dot product is positive, it belongs to the positive class, if it is negative it belongs to the negative class.
### Which one are important: Talking about the importance of each feature. Let's say the svm would find only one feature useful for separating the data, then the hyperplane would be orthogonal to that axis. So, we could say that the absolute size of the coefficient relative to the other ones gives an indication of how important the feature was for the separation. For example if only the first coordinate is used for separation, w will be of the form (x,0) where x is some non zero number and then |x|>0.
